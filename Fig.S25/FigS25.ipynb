```python
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
figS25 = pd.read_csv("figS25.csv")
sns.set_theme(style="ticks")#, rc={"axes.facecolor": (0, 0, 0, 0)})
colors = ["green", "darkorange", "magenta", "blue", "gray"]
sns.set_palette(sns.color_palette(colors))
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)

g = sns.catplot(
    data=figS25, aspect=.6,
    x="Inc type", y="Contig size (kbp)", jitter=False, linewidth=1,
    hue="Pi chart", col="Gene", col_order=["blaIMP-1", "blaIMP-6", "blaNDM-5"],
    hue_order=["Klebsiella spp", "Escherichia coli", "Enterobacter spp", "Citrobacter freundii", "Others"])#, palette="Dark2_r")
plt.xlabel("Inc type")
plt.ylabel("Contig size (bp)")

g.set_xticklabels(rotation=90)
g.fig.set_figheight(7)
g.fig.set_figwidth(14)
#g.set(ylabel='Plasmid size (bp)')
plt.savefig("catplot_hue_species_20220622.pdf", dpi=300, bbox_inches="tight")
```
